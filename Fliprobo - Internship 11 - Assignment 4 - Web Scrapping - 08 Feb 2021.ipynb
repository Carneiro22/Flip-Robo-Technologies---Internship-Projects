Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:

C) Contributors count
D) Language used

```python
# Connecting to the web driver
driver=webdriver.Chrome(r"C:\Users\Acer\Desktop\Internship 11 - Project 1 - Avinash Carneiro\chromedriver.exe")

# Specifying the url of the webpage
driver.get("https://github.com/")

# Scraping URL form the webpage
URL1 = []
urls = driver.find_elements_by_xpath('//a[@class="py-2 lh-condensed-ultra d-block link-gray-dark no-underline h5 Bump-link--hover"]')
for url in urls:
    URL1.append(url.get_attribute("href"))
print('\n')
    
print('The URL present on the Homepage are :')
print(URL1)
print('\n')

#Selecting the URL of explore category
url=URL1[1]
print('The URL of the Explore category is :')
print(url)
driver.get(url)
print('\n')

# Scraping URL form the webpage to reach Trnding category
URL2 = []
urls = driver.find_elements_by_xpath('//a[@data-selected-links="/trending /trending/developers /trending/developers /trending /trending"]')
for url in urls:
    URL2.append(url.get_attribute("href"))
    

# Selecting the URL of Trnding page
url2=URL2[0]
print('The URL of the Trending Repository Page is :')
print(url2)
driver.get(url2)

# Create empty lists
Repository=[]
Description=[]
Contributors=[]
Language=[]


# Getting the request for webpage server of https://github.com/trending.
page = requests.get(url2)
soup=BeautifulSoup(page.content, 'html.parser')

# To check Repository name
Repository_tags=soup.find_all('h1', class_='h3 lh-condensed')
for i in Repository_tags:  
    Repository.append(i.get_text().replace("\n", "").replace(" ", ""))  
# Selecting Top 10 trending Repositories Details
Repository=Repository[0:10]
Repository

# To check Description details
Description_tags=soup.find_all('p', class_='col-9 text-gray my-1 pr-4')
for i in Description_tags:  
    Description.append(i.get_text().replace("\n", ""))
# Selecting Top 10 trending Repositories Details
Description=Description[0:10]
Description

# To check Contributors details
Contributors_tags=soup.find_all('a', class_='muted-link d-inline-block mr-3')
for i in Contributors_tags:  
    Contributors.append(i.get_text().replace("\n", ""))

# Creating Dataframe for Contributors
df1=pd.DataFrame({})
df1['Count']=Contributors
# Removing Stars Ratings from the data
df1=df1.drop(df1.index[0:50:2])
df1.reset_index(drop=True, inplace=True)
# Selecting Top 10 trending Repositories Details
df1=df1[0:10]
df1

# To check Language used
Language_tags=soup.find_all('span', class_='d-inline-block ml-0 mr-3')
for i in Language_tags:  
    Language.append(i.get_text().replace("\n", ""))
# Selecting Top 10 trending Repositories Details
Language=Language[0:10]
Language

# Creating DataFrame for the Scrapped Data
pd.set_option('display.max_colwidth', None) 
Github_Trending=pd.DataFrame({})
Github_Trending['Repository Title']=Repository
Github_Trending['Repository Description']=Description
Github_Trending['Contributors Count']=df1
Github_Trending['Language Used']=Language

print('\n')
print('The Dataframe for details of Trending Repositories on https://github.com/ is :')
Github_Trending
```

```python
# Connecting to the web driver
driver=webdriver.Chrome(r"C:\Users\Acer\Desktop\Internship 11 - Project 1 - Avinash Carneiro\chromedriver.exe")

# Specifying the url of the webpage
driver.get("https://www.billboard.com/")

# Scraping URL form the webpage
URL1 = []
urls = driver.find_elements_by_xpath('//a[@class="header__subnav__link text--uppercase font--semi-bold"]')
for url in urls:
    URL1.append(url.get_attribute("href"))
    
# Selecting the "href" value which has Top 100 URL
URL=URL1[2]

# Selecting & clicking the current URL of Top 100 songs
driver.get(URL)
print('\n')
print('The URL of Top 100 songs is :', URL)

# Creating 4 empty lists
Songs=[]
Artist=[]
Last_Rank=[]
Peak_Rank=[]
Board_Week=[]

# Getting the request for webpage server:
page = requests.get(URL)
soup=BeautifulSoup(page.content, 'html.parser')

# To check Song name
songs_tags=soup.find_all('span', class_='chart-element__information__song text--truncate color--primary')
for i in songs_tags:  
    Songs.append(i.get_text().replace("\n", ""))
Songs

# To check Artist of the Song
artist_tags=soup.find_all('span', class_='chart-element__information__artist text--truncate color--secondary')
for i in artist_tags:  
    Artist.append(i.get_text().replace("\n", ""))
Artist

# To check Last_Rank of the song
last_tags=soup.find_all('span', class_='chart-element__meta text--center color--secondary text--last')
for i in last_tags:  
    Last_Rank.append(i.get_text().replace("\n", ""))
Last_Rank

# To check Peak_Rank of the song
peak_tags=soup.find_all('span', class_='chart-element__meta text--center color--secondary text--peak')
for i in peak_tags:  
    Peak_Rank.append(i.get_text().replace("\n", ""))
Peak_Rank

# To check Board_Week of the song
board_tags=soup.find_all('span', class_='chart-element__meta text--center color--secondary text--week')
for i in board_tags:  
    Board_Week.append(i.get_text().replace("\n", ""))
Board_Week


# Creating DataFrame for the Scrapped Data
BillBoard_Top_100=pd.DataFrame({})
BillBoard_Top_100['Song Name']=Songs
BillBoard_Top_100['Artist Name']=Artist
BillBoard_Top_100['Last week rank']=Last_Rank
BillBoard_Top_100['Peak Rank']=Peak_Rank
BillBoard_Top_100['Weeks on Board']=Board_Week

print('\n')
print('The Dataframe for details of Top 100 songs on https://www.billiboard.com/ is :')
BillBoard_Top_100
```

# Q 9

Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
# Connecting to the web driver
driver=webdriver.Chrome(r"C:\Users\Acer\Desktop\Internship 11 - Project 1 - Avinash Carneiro\chromedriver.exe")

# Specifying the url of the webpage
driver.get("https://www.imdb.com/list/ls095964455/")

#Lets create 4 empty lists
Name=[]
Year_Span=[]
Genre=[]
Run_Time=[]
Ratings=[]
Votes=[]

# Name of the TV series
for i in range(0, 1):
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']"):
        Name.append(i.text)
        
# Genre of the TV series
    for j in driver.find_elements_by_xpath("//span[@class='genre']"):
        Genre.append(j.text) 
        
# Run Time of TV series
    for k in driver.find_elements_by_xpath("//span[@class='runtime']"):
        Run_Time.append(k.text) 
        
# Ratings of the TV series
    for l in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']"):
        Ratings.append(l.text) 
        
# Votes of the TV series
    for m in driver.find_elements_by_xpath("//p[@class='text-muted text-small']"):
        Votes.append(m.text) 

# Year_Span of the TV series
    for n in driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']"):
        Year_Span.append(n.text)
        
        
# Creating DataFrame for Votes
df1=pd.DataFrame({})
df1['Votes_All']=Votes

# Finding the rows which has votes in it
df1=df1[df1['Votes_All'].str.contains("Votes")]
# Resetting the index to default after removing the required values
df1.reset_index(drop=True, inplace=True)
# Removing the data characters from the Votes column
df1['Votes_All']=df1['Votes_All'].str.replace('Votes:', '')

# Creating DataFrame for the Name
df1['Name']=Name
df1['Name']=df1['Name'].str.replace('\d+', '')
# Removing the unnecessary data/chracters from the Name column
df1['Name']=df1['Name'].apply(lambda x: x.replace('(','').replace(')','').replace('–',''))


# Creating DataFrame for the Scrapped Data
TV_Series=pd.DataFrame({})
TV_Series['Name']=df1['Name']
TV_Series['Year_Span']=Year_Span
TV_Series['Genre']=Genre
TV_Series['Run_Time']=Run_Time
TV_Series['Ratings']=Ratings
TV_Series['Votes']=df1['Votes_All']

print('\n')
print('The Dataframe for details of most watched tv series of all time from https://www.imdb.com/list/ls095964455/ is :')
TV_Series
```
