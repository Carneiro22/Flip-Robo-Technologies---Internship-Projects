```python
#So lets extract all the Genre name of books

Genre_tags=driver.find_elements_by_xpath("//p[@class='genre-links hidden-phone']")
Genre_tags[0:5]
```

```python
#So we will run a loop to literate over the tags extracted above and extract the text inside them.

for i in Genre_tags:
    
    BookGenre=i.text
    Genre.append(BookGenre)
Genre[0:5]
```

```python
page.content
```

```python
for i in review_tags:
    
    Book_Review.append(i.get_text().replace("\n", ""))
Book_Review[0:5]
```
